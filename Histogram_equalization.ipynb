```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image
image_path = '/content/drive/MyDrive/OpenCV/cute-and-small-dog-breed.jpg'
image = cv2.imread(image_path)
```

```python
# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Display histograms
plt.hist(image.ravel(), 256, [0, 256], color='r', alpha=0.5, label='RGB')
plt.hist(gray_image.ravel(), 256, [0, 256], color='k', alpha=0.5, label='Grayscale')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.show()
```
