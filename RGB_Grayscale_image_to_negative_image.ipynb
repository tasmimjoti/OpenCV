```python
import cv2
```

```python
def convert_to_negative(image):
    # Check if the image is grayscale
    if len(image.shape) == 2:  # Grayscale image
        negative_image = 255 - image
    else:  # RGB image
        negative_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        negative_image = cv2.bitwise_not(negative_image)

    return negative_image
```

```python
# Read the image
image_path = '/content/drive/MyDrive/OpenCV/cute-and-small-dog-breed.jpg'  # Change this to your image path
image = cv2.imread(image_path)
```

```python
negative_image = convert_to_negative(image)
output_path = '/content/drive/MyDrive/OpenCV/negative_image.jpg'  # Change this to your desired output path
cv2.imwrite(output_path, negative_image)
```
