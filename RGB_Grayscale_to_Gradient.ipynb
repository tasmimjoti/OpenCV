```python
import cv2
```

```python
def gradient_image(image):
    # Convert image to grayscale if it's not already
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) if len(image.shape) == 3 else image

    # Calculate gradient using Laplacian operator
    gradient_magnitude = cv2.Laplacian(gray_image, cv2.CV_8U)

    return gradient_magnitude
```

```python
# Read the image
image_path = '/content/drive/MyDrive/OpenCV/sunset-1373171_1280.jpg'  # Change this to your image path
image = cv2.imread(image_path)
```

```python
# Get the gradient image
gradient_img = gradient_image(image)
output_path = '/content/drive/MyDrive/OpenCV/gradient_img.jpg'  # Change this to your desired output path
cv2.imwrite(output_path, gradient_img)
```
